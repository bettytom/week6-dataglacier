```python
# Writing and reading files using YAML
import yaml
```

```python
%%writefile new_file.yaml
file_type: json
dataset_name: archive
file_name: news
inbound_delimiter: ","
columns:
  - title
  - description
  - reporter
  - time
  - category
  - images
```

```python
with open('new_file.yaml', 'r') as file:
    news = yaml.safe_load(file)
```

```python
# validate columns
def col_validate(data,config_file):
  yaml_columns=config_file['columns']
  if len(data.columns)== len(yaml_columns) and list(yaml_columns):
    print ("Column names and length are validated")
  else:
    print("column names and length validation has failed")
    missing_columns=list(set(data.columns).difference(yaml_columns))
    print("These columns are not in the file: ",missing_columns)
    missing_file_columns=list(set(yaml_columns).difference(data.columns))
    print("These columns are not in the file uploaded: ",missing_file_columns)
    print("data columns: ", data.columns)
    print("expected columns: ",yaml_columns)

```

```python
import pandas as pd
newfile=pd.read_json('news.json')
newfile
```

```python
#doing the action of the validation
col_validate(newfile,news)
```
